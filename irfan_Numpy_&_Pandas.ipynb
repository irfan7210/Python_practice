**Q1) Write a python program to count the frequency of unique values in NumPy array.**

```python
ini_array = np.array([10, 20, 5, 10, 8, 20, 8, 9])


# Get a tuple of unique values
# and their frequency in
# numpy array
dic = {}
for i in ini_array:
  if i in dic:
    dic[i] += 1
  else:
    dic[i] = 1

# print unique values array
print('\nprinting unique values in array')
# print(dic.values())
for key, val in dic.items():
  print(key, end = ' ')

# print frequency array
print('\nPrinting frequencies')
for key, val in dic.items():
  print(val, end = ' ')
```

array([2, 5])

**Q12) Compute the euclidean distance between two arrays a and b.**

```python
# Input
a = np.array([1,2,3,4,5])
b = np.array([4,5,6,7,8])
distance = np.linalg.norm(a - b)
print(distance)


# own logic
diff = b - a
sq_diff = diff**2
sum_of_squares = np.sum(sq_diff)
dist = np.sqrt(sum_of_squares)
print(dist)
```

**Q27) Write a Python program to get the latest number of confirmed, deaths, recovered and active cases of Novel Coronavirus (COVID-19) Country wise.**

```python
# Write your code here
import pandas as pd

covid_data = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/03-17-2020.csv')

# Group the data by Country/Region and get the latest numbers for each country
latest_data = covid_data.groupby('Country/Region').agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum'
}).reset_index()

# Calculate active cases by subtracting deaths and recovered from confirmed cases
latest_data['Active'] = latest_data['Confirmed'] - latest_data['Deaths'] - latest_data['Recovered']

# Display the latest data
print("Latest COVID-19 Data (as of 03-17-2020):\n")
print(latest_data[['Country/Region', 'Confirmed', 'Deaths', 'Recovered', 'Active']])

```
